### Marker


A format string consists of a part for color, marker and line.

'[marker][line][color]'

Each of them is optional. If not provided, the value from the style
cycle is used. Exception: If ``line`` is given, but no ``marker``,
the data will be a line without markers.

Other combinations such as ``[color][marker][line]`` are also
supported, but note that their parsing may be ambiguous.

**Markers**

Character| Description
--- | ---
``'.'`` | point marker
``','`` | pixel marker
``'o'`` | circle marker
``'v'`` | triangle_down marker
``'^'`` | triangle_up marker
``'<'`` | triangle_left marker
``'>'`` | triangle_right marker

**Line Styles**

Character| Description
--- | ---
``'-'`` | solid line style
``'--'``| dashed line style
``'-.'``| dash-dot line style
``':'`` | dotted line style

Example format strings:

    'b'    # blue markers with default shape
    'or'   # red circles
    '-g'   # green solid line
    '--'   # dashed line with default color
    '^k:'  # black triangle_up markers connected by a dotted line

**Colors**

The supported color abbreviations are the single letter codes

Character| Description
--- | ---
``'b'`` | blue
``'g'`` | green
``'r'`` | red
``'c'`` | cyan
``'m'`` | magenta
``'y'`` | yellow
``'k'`` | black
``'w'`` | white

### Linspace

In order to plot a function at a specific range $[from,to]$ we need to gather some points with some step. We could use range(from,to,step) or use the function np.linspace( from, to, step ) to divide to equal segments

```python
#from 0 to 50 divide to 10 equally spaced bins
np.linspace( 0, 10, 15 ).reshape( (-1,1) )
```

### Legend

To add labels to our lines we can specify the label attribute.

```python
x = np.linspace( -5, 5, 15 )
plt.plot( x, x, label='y=x' )
plt.plot( x, np.abs(x), label='y=|x|')
plt.legend()#make sure to make a call to legend, otherwise the box containing the labels won't appear
plt.show()
```

### Title - axes - alpha

```python
np.random.seed(1)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)#color per point
area = (30 * np.random.rand(N))**2#radius per point

plt.scatter(x, y, s=area, c=colors, alpha=0.5)#alpha value useful for overlapping

plt.xlabel( 'x' )
plt.ylabel( 'y' )

plt.title( 'Random alpha colorful circles with random radius')

plt.show()
```

### Multiple figures

Function subplot is defined as:
Number of Rows, Number of columns, Which Plot

There are different syntax, choose whoever you find convenient.

The 3 figures bellow are all equivalent.

Suplot just by specifying a number

```python
x = np.linspace( 0, 5, 50 )

plt.figure()
plt.subplot(211)
plt.plot( x, np.cos( 2*np.pi*x ), 'bo' )

plt.subplot(212)
plt.plot( x, np.cos(2*np.pi*x), 'r--')

plt.show()
```

Specify each argument seperated by ','

```python
x = np.linspace( 0, 5, 50 )

plt.figure()
plt.subplot(2,1,1)
plt.plot( x, np.cos( 2*np.pi*x ), 'bo' )

plt.subplot(2,1,2)
plt.plot( x, np.cos(2*np.pi*x), 'r--')

plt.show()
```

By axes

```python
x = np.linspace( 0, 5, 50 )

fig, (axes) = plt.subplots( 2, 1 )

fig.suptitle('Subplots referred with array')#

axes[0].plot( x, np.cos( 2*np.pi*x ), 'bo' )
axes[0].set_ylabel('First')

axes[1].plot( x, np.cos(2*np.pi*x), 'r--')
axes[1].set_ylabel( 'Second' )             
axes[1].set_xlabel('x')

plt.show()
```

### 4x4

```python
x = np.linspace( 0, 5, 50 )

plt.figure( figsize=(20,5) )
plt.subplot(2,2,1)
plt.plot( x, np.sin( 2*np.pi*x ), 'bo' )
plt.ylabel( 'sin')

plt.subplot(2,2,2)
plt.plot( x, np.cos(2*np.pi*x), 'r--')
plt.ylabel( 'cos' )

plt.subplot(2,2,3)
plt.plot( x, -np.sin(2*np.pi*x), 'mo' )
plt.ylabel( '-sin' )

plt.subplot(2,2,4)
plt.plot( x, -np.cos(2*np.pi*x), 'g--')
plt.ylabel( '-cos' )

plt.show()
```
