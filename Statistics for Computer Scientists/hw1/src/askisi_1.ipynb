```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Provided data
data1 = [30.3, 31.0, 31.2, 32.1, 32.6, 32.7, 33.4, 33.6, 34.2, 34.5]
data2 = [0.0, 0.0, 0.2, 0.8, 1.2, 1.4, 3.2, 4.2, 6.4, 9.0]
data3 = [0, 1, 6, 8, 10, 13, 15, 16, 17, 17, 18, 18, 20, 20, 21, 25, 26, 30, 35, 39, 40,
41, 43, 44, 46, 48, 52, 54, 58, 59, 59, 60, 66, 81, 86, 87, 88, 89, 94, 96]
```

```python
# Convert data from lists to numpy arrays
data1 = np.array(data1)
data2 = np.array(data2)
data3 = np.array(data3)

print("DATA 1 is: ", data1)
print("***************************************************")
print("DATA 2 is: ", data2)
print("***************************************************")
print("DATA 3 is: ", data3)
print("***************************************************")
```

```python
def createStemplot(data, dim=1):
    # Create stem and leaf components
    if dim == 1:
        stem = [int(x // 1) for x in data]
        leaf = [int(round((x % 1) * 10)) for x in data]
        plt.figure(figsize=(5, 8))
        
    elif dim == 2:
        stem = [int(x // 10) for x in data]
        leaf = [int((x % 10)) for x in data]
        plt.figure(figsize=(6, 10))
        
    # Create the stemplot using matplotlib

    plt.stem(stem, leaf, linefmt='-', markerfmt='o', basefmt=" ")
    plt.yticks(np.arange(min(leaf), max(leaf)+1, 1))
    plt.xticks(np.arange(min(stem), max(stem)+1, 1))
    plt.xlabel('Stem')
    plt.ylabel('Leaf')
    plt.title('Stemplot')
    plt.grid(True)
    
    plt.show()
```

```python
createStemplot(data1)
createStemplot(data2)
createStemplot(data3, 2)
```

```python
def createBoxplot(data):
    plt.figure(figsize=(5, 8))
    plt.boxplot(data)
    plt.title('Boxplot')
    plt.show()
```

```python
createBoxplot(data1)
createBoxplot(data2)
createBoxplot(data3)
```

```python
# Function to plot density alongside real data points
def plot_density(data, mean, std, title):
    # Generate synthetic normal distribution data
    x = np.linspace(min(data) - 2, max(data) + 2, 1000)
    y = stats.norm.pdf(x, mean, std)
    
    # Plotting
    plt.figure(figsize=(10, 4))
    plt.title("Histogram and Density Curve for " + title)
    plt.suptitle(" (Mean = " + str(mean.round(4)) + ", Standard Deviation = " + str(std.round(4)) + ")")
    plt.hist(data, bins='auto', density=True, alpha=0.6, label='Real Data', color='darkgreen')
    plt.plot(x, y, 'k', linewidth=2, label='Normal Distribution')
    plt.legend()
    plt.show()
```

```python
# Plotting density plots for each dataset
plot_density(data1, mean_data1, std_data1, 'Data1')
plot_density(data2, mean_data2, std_data2, 'Data2')
plot_density(data3, mean_data3, std_data3, 'Data3')
```

```python

# Function that plots the QQ plot for a given dataset
def plot_qq(data, title):
    plt.figure(figsize=(10, 4))
    stats.probplot(data, plot=plt)
    plt.title("QQ Plot for " + title)
    plt.show()
```

```python
# Plotting QQ plots for each dataset
plot_qq(data1, 'Data1')
plot_qq(data2, 'Data2')
plot_qq(data3, 'Data3')
```
