```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

### Data Provided


```python
data = np.array([[5.7, 4.6, 6.4, 6.3, 6.9, 5.2, 4.9, 5.4, 4.9, 5.6, 5.4, 5.3, 4.9, 5.1, 5.0, 6.0, 6.3, 5.4, 5.3, 5.4]])
n = len(data) # sample size = 20
df = pd.DataFrame(data.T, columns=['data'])
df.head()
```

#### a) Τα δεδομένα είναι κατάλληλα για τις μεθόδους συμπερασματολογίας που γνωρίζουμε;

```python
sns.barplot(data=df['data'])
plt.show()
```

```python
# Creating a Q-Q plot
plt.figure(figsize=(8, 6))
stats.probplot(df['data'], dist="norm", plot=plt)
plt.title('Q-Q Plot for Normality Check')
plt.show()
```

#### b). Βρείτε τη μέση τιμή και τυπική απόκλιση για τα δεδομένα αυτά.

```python
mean = df.mean()[0]
mean
```

```python
std = df.std(ddof=1)[0] # ddof=1 for sample standard deviation
std
```

#### c). Εκτιμήστε τη μέση τιμή 𝜇 της απόδοσης του αυτοκινήτου, με ένα 95% διάστημα εμπιστοσύνης χρησιμοποιώντας τα παραπάνω δεδομένα.

```python
sem = stats.sem(df)[0]
sem
```

```python
confidence_level = 0.95
t_score_95 = stats.t.ppf( (1 + confidence_level) / 2, n-1 )
t_score_95
```

```python
marging_of_error = sem * t_score_95
marging_of_error
```

```python
confidence_interval_95 = (mean - abs(marging_of_error), mean + abs(marging_of_error))
confidence_interval_95
```

### Τρόπος με μία γραμμή κώδικα.

```python
confidence_interval_95 = stats.t.interval(confidence=0.95, df=len(df)-1, loc=mean, scale=sem) 
confidence_interval_95
```
