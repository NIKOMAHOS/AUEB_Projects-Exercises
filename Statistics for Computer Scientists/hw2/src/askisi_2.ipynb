#### a. Λαμβάνεται ένα τυχαίο δείγμα μεγέθους 20 από πληθυσμό με τυπική απόκλιση 12. Η τυπική απόκλιση του δειγματικού μέσου είναι 12/20.

#### b. Ένας ερευνητής χρησιμοποιεί σε έναν έλεγχο σημαντικότητας τη μηδενική υπόθεση 𝐻0: 𝑥̅= 10.
- Είναι ασυνήθιστο για μία μηδενική υπόθεση να αναφέρεται σε δειγματική παράμετρο. Συνήθως, μία μηδενική υπόθεση αφορά κάποια παράμετρο σχετική με τον πληθυσμό. Στην συγκεκριμένη περίπτωση θα ήταν πιο αποδεκτό η μηδενική υπόθεση να ήταν η ακόλουθη: μ=10, δεδομένου ότι μ η μέση τιμή του πληθυσμού.


#### c. Σε μια στατιστική έρευνα με 𝑥̅= 45 απορρίπτεται η μηδενική υπόθεση 𝛨0: 𝜇 = 54 όταν η εναλλακτική είναι 𝐻𝑎: 𝜇 > 54.
- Η μηδενική υπόθεση μ=54 απορρίπτεται και επιλέγεται η εναλλακτική υπόθεση 
    μ>54, ενώ ισχύει ότι ο δειγματικός μέσος 𝑥̅ είναι μικρότερος από την υποτιθέμενη μέση τιμή (του πληθυσμού) [𝑥̅=45 < 54]. Αυτό δεν είναι σωστό, διότι για να αποδεχτούμε την εναλλακτική υπόθεση (Ha: μ>54) θα πρέπει να ισχύει μόνο αν ο δειγματικός μέσος είναι αρκετά μεγαλύτερος του 54 και σίγουρα όχι αν είναι μικρότερος.


#### d. Σε μια στατιστική έρευνα όπου p value = 0.52 απορρίπτεται η μηδενική υπόθεση.
- Η απόρριψη της μηδενικής υπόθεσης με p-value = 0.52 είναι λανθασμένη. Γενικά μία p-value μεγαλύτερη από το επίπεδο σημαντικότητας (συνήθως 0.05) δεν δύναται να μας οδηγήσει στην απόρριψη της μηδενικής υπόθεσης. Απορρίπτουμε την μηδενική υπόθεση αν η p-value είναι μικρότερη από το επίπεδο σημαντικότητας (πχ. 0.05) και όχι όταν είναι υψηλότερη (όπως στην συγκεκριμένη περίπτωση που είναι 0.52)
